```python
def get_averaged_result(data):
    N = len(data)
    averaged_result = np.zeros(32)
    shots = 4000
    for data_num in range(N):
        for i in range(32):
            result = int(bin(i)[2:])
            result = "{:05d}".format(result)
            averaged_result[i] += data[data_num][result]

    averaged_result = averaged_result / (N*shots)
    return averaged_result
```

```python
def normalize_data(data):
    N = len(data)
    normalized_data = []
    for data_num in range(N):
        this_prob_sum = 0
        this_dict = {}
        temp = data[data_num]
        for i in range(32):
            result = int(bin(i)[2:])
            result = "{:05d}".format(result)

            if result in temp.keys():
                this_prob_sum += data[data_num][result]
        for i in range(32):
            result = int(bin(i)[2:])
            result = "{:05d}".format(result)
            if result in temp.keys():
                this_dict[result] = data[data_num][result] / this_prob_sum
        normalized_data.append(this_dict)
    return normalized_data
```
